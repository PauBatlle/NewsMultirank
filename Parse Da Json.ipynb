```python
M = []
for i in t(range(len(data11)+len(data21))):
    s = []
    if (i < len(data11)):
        s.append(data12[i]['postText'][0])
        s.append(data11[i]['truthMean'])
        assert(data12[i]['id'] == data11[i]['id'])
    else:
        s.append(data22[i-len(data11)]['postText'][0])
        idd = data22[i-len(data11)]['id']
        for j in range(len(data22)):
            if data21[j]['id'] == idd:
                s.append(data21[j]['truthMean'])
                break
    M.append(s)
```
